```python
def plot_2d_slices(image, label, slice_no):
    fig, axs = plt.subplot_mosaic(
        [
            ['img1', 'lab1'], 
            ['img2', 'lab2'], 
        ],
         layout='constrained')


    axs['img1'].imshow(image[:, :, slice_no ], cmap="gray")
    axs['lab1'].imshow(label[:, :, slice_no ])
    axs['img2'].imshow(image[:, :, slice_no + 15], cmap="gray")
    axs['lab2'].imshow(label[:, :, slice_no + 15])
    plt.show()

d_dict = loader(data_dict[0])
slice = 210
image, label = d_dict["image"], d_dict["label"]

plot_2d_slices(image, label, slice)
```
