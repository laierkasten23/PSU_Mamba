```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# Data for experiments
experiments = ['Exp 1: T1', 'Exp 2: FLAIR', 'Exp 3: T1xFLAIR', 'Exp 4: T1+FLAIR']

# Data for losses
aschoplex_loss = [0.25, 0.28, 0.18, 0.22]
phusegplex_loss = [0.23, 0.26, 0.16, 0.19]
umamba_loss = [0.22, 0.24, 0.14, 0.18]

# Data for Dice scores
aschoplex_dice = [0.78, 0.76, 0.85, 0.80]
phusegplex_dice = [0.80, 0.78, 0.87, 0.83]
umamba_dice = [0.83, 0.81, 0.89, 0.85]

# Data for IoU scores
aschoplex_iou = [0.65, 0.63, 0.72, 0.68]
phusegplex_iou = [0.67, 0.65, 0.75, 0.71]
umamba_iou = [0.70, 0.68, 0.77, 0.73]

# 2. Bar Graph Example (Dice Scores)
plt.figure(figsize=(10,6))
x = np.arange(len(experiments))
width = 0.2

plt.bar(x - width, aschoplex_dice, width, label='ASCHOPLEX', color='b')
plt.bar(x, phusegplex_dice, width, label='PHUSegPlex', color='g')
plt.bar(x + width, umamba_dice, width, label='UMAMBA', color='r')

plt.xlabel('Experiments')
plt.ylabel('Dice Score')
plt.title('Comparison of Dice Scores Across Experiments and Models')
plt.xticks(x, experiments)
plt.legend()
plt.show()

# 3. Line Graph Example (Loss)
plt.figure(figsize=(10,6))
plt.plot(experiments, aschoplex_loss, marker='o', label='ASCHOPLEX', color='b')
plt.plot(experiments, phusegplex_loss, marker='o', label='PHUSegPlex', color='g')
plt.plot(experiments, umamba_loss, marker='o', label='UMAMBA', color='r')

plt.xlabel('Experiments')
plt.ylabel('Loss')
plt.title('Loss Across Experiments and Models')
plt.legend()
plt.show()

# 4. Heatmap Example (Dice Scores)
plt.figure(figsize=(8,6))
data = np.array([aschoplex_dice, phusegplex_dice, umamba_dice])
sns.heatmap(data, annot=True, cmap="YlGnBu", xticklabels=experiments, yticklabels=['ASCHOPLEX', 'PHUSegPlex', 'UMAMBA'])
plt.title('Dice Score Heatmap Across Models and Experiments')
plt.show()

# 5. Radar Chart (Spider Plot) for One Experiment
from math import pi

labels = ['Loss', 'Dice Score', 'IoU']
aschoplex = [0.25, 0.78, 0.65]
phusegplex = [0.23, 0.80, 0.67]
umamba = [0.22, 0.83, 0.70]

# Number of variables
num_vars = len(labels)

# Set up the radar chart
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
aschoplex += aschoplex[:1]  # Closing the radar chart
phusegplex += phusegplex[:1]
umamba += umamba[:1]
angles += angles[:1]

# Create Radar Chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
ax.fill(angles, aschoplex, color='b', alpha=0.25)
ax.fill(angles, phusegplex, color='g', alpha=0.25)
ax.fill(angles, umamba, color='r', alpha=0.25)

ax.plot(angles, aschoplex, color='b', linewidth=2, label='ASCHOPLEX')
ax.plot(angles, phusegplex, color='g', linewidth=2, label='PHUSegPlex')
ax.plot(angles, umamba, color='r', linewidth=2, label='UMAMBA')

ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

plt.title("Radar Chart for Exp 1 (T1)")
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))
plt.show()
```

```python
import scipy.stats as stats

# Perform statistical test (e.g., ANOVA) to check if there are significant differences in Dice scores between models
_, p_value_dice = stats.f_oneway(aschoplex_dice, phusegplex_dice, umamba_dice)

# Adding whiskers to represent standard deviation
aschoplex_dice_std = [0.02, 0.03, 0.01, 0.02]  # Standard deviation example for ASCHOPLEX
phusegplex_dice_std = [0.01, 0.02, 0.01, 0.02]  # Standard deviation example for PHUSegPlex
umamba_dice_std = [0.01, 0.02, 0.01, 0.02]  # Standard deviation example for UMAMBA

# 2. Bar Graph with Whiskers Example (Dice Scores + Standard Deviation)
plt.figure(figsize=(10,6))
x = np.arange(len(experiments))
width = 0.2

plt.bar(x - width, aschoplex_dice, width, yerr=aschoplex_dice_std, label='ASCHOPLEX', color='b', capsize=5)
plt.bar(x, phusegplex_dice, width, yerr=phusegplex_dice_std, label='PHUSegPlex', color='g', capsize=5)
plt.bar(x + width, umamba_dice, width, yerr=umamba_dice_std, label='UMAMBA', color='r', capsize=5)

plt.xlabel('Experiments')
plt.ylabel('Dice Score')
plt.title(f'Comparison of Dice Scores Across Experiments and Models (p-value: {p_value_dice:.4f})')
plt.xticks(x, experiments)
plt.legend()
plt.show()
```
