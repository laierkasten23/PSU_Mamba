```python
import os
import nibabel as nib
import matplotlib.pyplot as plt

def plot_subject_slices(subject_path, direction, slice_index):
    # File pattern
    file_patterns = {
        'T1': '_T1.nii',
        'FLAIR': '_FLAIR.nii',
        'T1xFLAIR': '_T1xFLAIR.nii',
        'T1_mask': '_ChP_mask_T1_manual_seg.nii',
        'FLAIR_mask': '_ChP_mask_FLAIR_manual_seg.nii',
        'T1xFLAIR_mask': '_ChP_mask_T1xFLAIR_manual_seg.nii'
    }
    
    # Extract subject identifier (XXX) from the path
    subject_id = os.path.basename(subject_path)
    
    # Load images
    images = {}
    for key, pattern in file_patterns.items():
        image_path = os.path.join(subject_path, subject_id + pattern)
        images[key] = nib.load(image_path).get_fdata()
    
    # Select the appropriate slice based on the direction
    if direction == 'x':
        slices = {key: img[slice_index, :, :] for key, img in images.items()}
    elif direction == 'y':
        slices = {key: img[:, slice_index, :] for key, img in images.items()}
    elif direction == 'z':
        slices = {key: img[:, :, slice_index] for key, img in images.items()}
    else:
        raise ValueError("Direction must be 'x', 'y', or 'z'")
    
    # Plot the images
    fig, axs = plt.subplots(2, 3, figsize=(8, 6))
    
    # First row: T1, FLAIR, T1xFLAIR
    axs[0, 0].imshow(slices['T1'].T, cmap='gray', origin='lower')
    axs[0, 0].set_title('T1 Image')
    axs[0, 0].axis('off')
    
    axs[0, 1].imshow(slices['FLAIR'].T, cmap='gray', origin='lower')
    axs[0, 1].set_title('FLAIR Image')
    axs[0, 1].axis('off')
    
    axs[0, 2].imshow(slices['T1xFLAIR'].T, cmap='gray', origin='lower')
    axs[0, 2].set_title('T1xFLAIR Image')
    axs[0, 2].axis('off')
    
    # Second row: T1 Mask, FLAIR Mask, T1xFLAIR Mask
    axs[1, 0].imshow(slices['T1_mask'].T, cmap='gray', origin='lower')
    axs[1, 0].set_title('T1 Mask')
    axs[1, 0].axis('off')
    
    axs[1, 1].imshow(slices['FLAIR_mask'].T, cmap='gray', origin='lower')
    axs[1, 1].set_title('FLAIR Mask')
    axs[1, 1].axis('off')
    
    axs[1, 2].imshow(slices['T1xFLAIR_mask'].T, cmap='gray', origin='lower')
    axs[1, 2].set_title('T1xFLAIR Mask')
    axs[1, 2].axis('off')
    
    plt.tight_layout()
    plt.show()
```

```python

# Example usage
subject_path = '/home/linuxuser/user/data/Final_coreg_240829/022'
#subject_path = '/home/linuxuser/user/data/pazienti_final_tests/089'    
plot_subject_slices(subject_path, 'z', 120) #130 (120 (T1)- 149 (product))
```
